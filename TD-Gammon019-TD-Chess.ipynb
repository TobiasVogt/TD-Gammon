```python
class ChessModelPlayer():
    
    def __init__(self, player, model):
        self.player = player
        self.model = model

    def get_action(self, actions, game):
        # Variablen initialisieren
        best_value = float("-inf")
        best_action = None
        # Alle Zge durchsuchen
        for a in actions:
            # Zug ausfhren
            game.push(a)
            # Spielstatus bewerten
            value = self.get_value(game, self.player)
            # Besten merken
            if value > best_value:
                best_value = value
                best_action = a
            # Spiel zurcksetzen
            game.pop()
        return best_action
    
    def get_value(self, game, player):
        features = game.extractFeatures(player)
        v = self.model.get_output(features)
        v = 1 - v if self.player == chess.BLACK else v
        return v
    
    def get_name(self):
        return "TD-ChessPlayer"
```

```python
import numpy as np
import chess
import random

class ChessGame():
    
    def __init__(self):
        self.board = chess.Board()
        
    def push(self, a):
        self.board.push(a)
        
    def pop(self):
        self.board.pop()
        
    def extractFeatures(self, player):
        features = []
        for square in chess.SQUARES:
            p = self.board.piece_at(square)
            if p and p.color == chess.WHITE:
                features.append(p.piece_type / 6)
            else:
                features.append(0)
        for square in chess.SQUARES:
            p = self.board.piece_at(square)
            if p and p.color == chess.BLACK:
                features.append(p.piece_type / 6)
            else:
                features.append(0)
        if player == chess.BLACK:
            features += [1, 0]
        else:
            features += [0, 1]
        return np.array(features).reshape(1, -1)
    
    def play(self, player, debug=False):
        #Wer anfngt ist zufllig
        active_player = bool(random.getrandbits(1))
        #Solange spielen bis es einen Gewinner gibt
        while not self.board.is_game_over():
            #Zug ausfhren
            self.next_step(player[active_player], debug=debug)
            #Der andere Spieler ist dran
            active_player = not active_player
        #Siegesstats ausgeben
        return self.board.result()

    def next_step(self, player, debug=False):
        #Wrfeln
        roll = (random.randint(1,6), random.randint(1,6))
        #Zge berechnen
        moves = [m for m in self.board.legal_moves]
        #Spieler fragen welche der Zge er gerne ausfhren mchte
        move = player.get_action(moves, self) if moves else None
        #Zug ausfhren falls es mglich ist 
        self.board.push(move)
        #Debuggen
        if debug:
            print(self.board)
        
```

```python
from NeuralNetModel import TDGammonModel
from Player import RandomPlayer

class TDChessModel(TDGammonModel):
    
    def test(self, enemyPlayer=RandomPlayer(chess.WHITE), games=100, debug=False):
        players = [ChessModelPlayer(chess.BLACK, self), enemyPlayer]
        
        for i in range(games):
            game = ChessGame()

            winner = game.play(players, debug=debug)
            print(winner)
            
    def train(self, games, validation_interval, test_games=10):
        #Selbsttraining, Modell vs Modell
        players = [ChessModelPlayer(chess.BLACK, self), ChessModelPlayer(chess.WHITE, self)]

        for i in range(games):
            #Immer wieder zwischendurch testen und den Fortschritt speichern
            if i != 0 and i % validation_interval == 0:
                self.saver.save(self.sess, self.checkpoint_path + 'checkpoint.ckpt', global_step=global_step)
                print("Progress saved!")
                self.test(games = test_games)

            #Spiel initialisieren
            game = ChessGame()
            player_num = bool(random.getrandbits(1))
            
            x = game.extractFeatures(players[player_num].player)

            #Spiel spielen bis es einen Sieger gibt
            game_step = 0
            while not game.board.is_game_over():
                game.next_step(players[player_num])
                player_num = not player_num

                #Das Modell mit jeden Schritt im Spiel trainieren
                x_next = game.extractFeatures(players[player_num].player)
                V_next = self.get_output(x_next)
                
                self.sess.run([self.train_op, self.delta_op], feed_dict={ self.x: x, self.V_next: V_next })

                x = x_next
                game_step += 1

            #Gewinner ermitteln
            winner = 0.5
            if game.board.result() == "1-0":
                winner = 1
            elif game.board.result() == "0-1":
                winner = 0

            #Zu guter letzt reinforcement learning: Dem Modell noch eine "Belohnung" geben, wenn es gewonnen hat
            _, global_step = self.sess.run([
                self.train_op,
                self.global_step,
            ], feed_dict={ self.x: x, self.V_next: np.array([[winner]], dtype='float') })

            #Konsolenausgabe hbsch aufbereiten
            print("Game %d/%d (%s) in %d turns" % (i, games, game.board.result(), game_step))
        #Am Ende noch mal speichern und 100 testen!
        self.saver.save(self.sess, self.checkpoint_path + 'checkpoint.ckpt', global_step=global_step) 
        self.test(games = test_games)
        
    def get_name(self):
        return "TD-Chess"
```
